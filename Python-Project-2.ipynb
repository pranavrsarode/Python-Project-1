```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df
```

```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
```

```python
df
```

```python
total_sales=df['Sales'].sum()
print('Total Sales=',total_sales)
```

```python
total_profit=df['Profit'].sum()
print('Total Profit=',total_profit)
```

```python
total_average=df['Sales'].mean()
print('Average Sales',total_average)
```

```python
region_sales=df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
category_sales=df.groupby('Category')['Sales'].sum()
print(category_sales)
```

```python
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title('Category Wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x=region_sales.index,y=region_sales.values)
plt.title("Region Wise Sales")
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
top_products=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
top_products.plot(kind='line', marker='o')
plt.title("Top 10 Selling Products")
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
monthly_sales=df.groupby('Month')['Sales'].sum().sort_values(ascending=False)
print(monthly_sales)
```

```python
monthly_sales.plot(kind='line',marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
region_profit=df.groupby('Region')['Profit'].sum()
print(region_profit)
```

```python
region_profit.plot(kind='pie', autopct='%1.1f%%')
plt.title("Region Wise Profit Distribution")
plt.xlabel('Region')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
top_customers=df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
top_customers.plot(kind='bar')
plt.title("Top 10 Customers")
plt.xlabel('Customers Name')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

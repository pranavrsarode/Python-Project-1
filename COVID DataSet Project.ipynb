```python
total_cases = df['cases'].sum()
print(total_cases)
```

```python
total_deaths = df['deaths'].sum()
print(total_deaths)
```

```python
top_states = df.groupby('state')['cases'].sum().sort_values(ascending = False).head(10)
print(top_states)
```

```python
top_states.plot(kind='bar')
plt.title("Top 10 States by COVID Cases")
plt.xlabel("State")
plt.ylabel("Cases")
plt.xticks(rotation=35)
plt.show()
```

```python
top_deaths = df.groupby('state')['deaths'].sum().sort_values(ascending = False).head(10)
print(top_states)
```

```python
top_deaths.plot(kind='bar')
plt.title("Top 10 States by Deaths")
plt.xlabel("States")
plt.ylabel("Deaths")
plt.xticks(rotation=35)
plt.show()
```

```python
daily_cases=df.groupby('date')['cases'].sum()
print(daily_cases)
```

```python
plt.figure(figsize=(12,6))
plt.plot(daily_cases)
plt.title("COVID Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.grid(True)
plt.show()
```

```python
sns.heatmap(df[['cases','deaths']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

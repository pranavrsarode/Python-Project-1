```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Python Projects\archive\Diwali Sales Data.csv", encoding='latin1')
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df
```

```python
df.dtypes
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of Customers By Gender")
plt.show()
```

```python
gender_sales=df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
print(gender_sales)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
plt.title('Sales by Age Group')
plt.show()
```

```python
orders_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=orders_state,x='State',y='Orders')
plt.title("Top 10 States By Orders")
plt.show()
```

```python
orders_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=orders_state,x='State',y='Amount')
plt.title("Top 10 States By Amount")
plt.show()
```

```python
ax = sns.countplot(data=df, x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Marital Status Count')
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
plt.title('Sales by Marital Status')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Occupation Status Count')
plt.show()
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.lineplot(data=sales_occ,x='Occupation',y='Amount')
plt.title('Occupation Status By Amount')
plt.show()
```

```python
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Product_Category Count')
plt.xticks(rotation=35)
plt.show()
```

```python
sales_category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.lineplot(data=sales_category,x='Product_Category',y='Amount')
plt.title("Sales By Product_Category")
plt.xticks(rotation=35)
plt.show()
```

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(data=sales_product,x='Product_ID',y='Orders')
plt.title("Top 10 Selling Products")
plt.show()
```
